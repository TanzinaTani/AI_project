```python
mail=prob['ham_or_spam']
sum_ham=0
sum_spam=0
for iteam in mail:
    if (iteam=='ham'):
         sum_ham+=1
    elif (iteam=='spam'):
        sum_spam+=1
print('total ham is: ',sum_ham)
print('total spam is: ',sum_spam)
prob_ham= sum_ham/(sum_ham+sum_spam)
prob_spam=sum_spam/(sum_ham+sum_spam)
print("probability of ham is: ",prob_ham)
print("probability of spam is: ",prob_spam)
```

```python
p_ham = prob_ham
p_spam = prob_spam
```

```python
prob.shape
```

```python
accurate_prediction = 0
wrong_prediction = 0
for i in range(0, 1000):
    sample = dft3.loc[i]
    #applying gaussian
    probability_ham = p_ham
    probability_spam = p_spam
    #calculationg the probability of the sample being ham
    for item in list1:        
        try:
            x = sample.loc[item]
            #print(x)
            mean_x = mean.loc[0, item]
            variance_x = variance.loc[0, item]
            individual_item_probability = (1.0/math.sqrt(2*3.1416*variance_x))*math.exp(-((x - mean_x)*(x - mean_x))/(2*variance_x))
            probability_ham *= individual_item_probability
        except:
            #print("exception!")
            probability_ham *= 0.5
        
    #calculationg the probability of the sample being spam
    for item in list1:        
        try:
            x = sample.loc[item]
            #print(x)
            mean_x = mean.loc[1, item]
            variance_x = variance.loc[1, item]
            individual_item_probability = (1.0/math.sqrt(2*3.1416*variance_x))*math.exp(-((x - mean_x)*(x - mean_x))/(2*variance_x))
            probability_spam *= individual_item_probability
        except:
            #print("exception!")
            probability_spam *= 0.5
            
    result = ''
    if probability_ham >= probability_spam:
        result = 'ham'
        #print('ham')
    else:
        result = 'spam'
        #print('spam')
    #calculating accuracy
    if sample.ham_or_spam == result:
        accurate_prediction += 1
    else:
        wrong_prediction += 1

print(accurate_prediction)
print(wrong_prediction)
accuracy = (accurate_prediction/(accurate_prediction+wrong_prediction))*100
print(f'{accuracy}%')
           
```
